# finding-missing-person-using-AI

```python
import keras
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen=ImageDataGenerator(rescale=1./255,
                                shear_range=0.2,
                                rotation_range=180,
                                horizontal_flip=True,
                                zoom_range=0.2)
test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
x_train=train_datagen.flow_from_directory(r'C:\Users\admin\Downloads\Dataset\Dataset\trainset',
                                          target_size=(64,64),batch_size=32,
                                          class_mode='binary')
```

```python
x_test=train_datagen.flow_from_directory(r'C:\Users\admin\Downloads\Dataset\Dataset\testset', 
                                         target_size=(64,64),batch_size=32,
                                         class_mode='binary')
```

```python
print(x_train.class_indices)
```

adding models and layers

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPool2D
from tensorflow.keras.layers import Flatten
```

```python
model=Sequential()
```

```python
model.add(Conv2D( 32,3,3,input_shape=(64,64,3),activation='relu'))
```

```python
model.add(MaxPool2D(pool_size=(2,2)))
```

```python
model.add(Flatten())
```

```python
model.summary()
```

```python
model.add(Dense(units=128,kernel_initializer='random_uniform',activation='relu')
```

```python
model.add(Dense(units=128,kernel_initializer='random_uniform',activation='relu'))
```

```python
model.add(Dense(units=1,activation='sigmoid',kernel_initializer='random_uniform'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit_generator(x_train , steps_per_epoch=8 , epochs=300, validation_data = x_test, validation_steps = 2 )
```

save the model

```python
model.save('Missing_1.h5')
```
